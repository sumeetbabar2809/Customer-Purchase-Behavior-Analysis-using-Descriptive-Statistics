## Task 3 - Probability Distributions üé≤

- **Identify variables that could follow specific probability distributions (e.g., Binomial, Normal). Calculate probabilities and expected values based on these distributions.**



**Deliverables**:

- **Determination of suitable probability distributions for relevant variables and corresponding calculated probabilities and expected values.**:

  

```python
total_purchases = lambda data: data[['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']].sum(axis=1)

# Lambda function to calculate expected_purchases
expected_purchases = lambda data: total_purchases(data).mean()
# Calculating p_purchase in one line using lambda and method chaining
probab_purchase = (lambda data: total_purchases(data).mean() / total_purchases(data).max())(data)

from scipy.stats import norm
x_income, y_income = data['Income'].mean(), data['Income'].std()
prob_income_gt_70000 = 1 - norm.cdf(70000, x_income, y_income)
expected_income = x_income + y_income * norm.ppf(prob_income_gt_70000)



# Printing results
print(f"Expected Purchases: {expected_purchases(data):.2f}")
print(f"Probability of Purchase: {probab_purchase:.4f}")
print("-"*180)

print(f"Probability of income > $70,000: {prob_income_gt_70000:.4f}")
print(f"Expected income: {expected_income:.2f}")
```

**Expected Purchases:** Customers are expected to make an average of 14.86 purchases across different categories.

**Probability of Purchase:** There's a 33.78% likelihood that customers will make a purchase based on historical data.

**Probability of Income > 70,000 dollars:** Approximately 23.92% of individuals in the dataset have an income greater than 70,000 dollars.

**Expected Income:** The average estimated income for individuals in the dataset is 34,494.50 dollars.


These insights inform strategies for marketing, customer segmentation, and operational planning based on customer behavior and income distribution.

```python
# Calculate TotalSpend as the sum of various purchase types
data['TotalSpend'] = data[['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']].sum(axis=1)

# Plotting with Seaborn using an orange gradient
plt.figure(figsize=(15, 5))
sns.scatterplot(x='Income', y='TotalSpend', data=data, hue='Income', color='Red', legend=True)
plt.title('Relationship between Income and Total Spending')
plt.xlabel('Income')
plt.ylabel('Total Spending')
plt.show()
```

**Relationship between Income and Total Spending:**

* The scatter plot illustrates the relationship between income levels (**Income**) and total spending (**Total Spending**).
*Higher income levels are represented by darker shades of orange in the plot, reflecting increased total spending.
*This visualization provides insights into how income influences spending behavior across various purchase categories in your dataset.

```python
numerical_cols = data.select_dtypes(include=['number']).columns

# Calculate correlation matrix for numerical columns
correlation_matrix = data[numerical_cols].corr()

# Visualize correlation matrix using a heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```
